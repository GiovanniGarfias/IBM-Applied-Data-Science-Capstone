We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,


```python
from IPython.core.display import HTML, display
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)

display(HTML(site_map._repr_html_()))
```

```python
# Task 2: Mark the success/failed launches for each site on the map
# 2) Add per-launch markers with MarkerCluster (simple & compact)
from folium.plugins import MarkerCluster
from folium.features import DivIcon
from IPython.core.display import HTML, display

# create a cluster for all launches
launch_cluster = MarkerCluster(name='All Launches').add_to(site_map)

# helper for marker color
def color_by_class(c):
    try:
        return 'green' if int(c) == 1 else 'red'
    except:
        return 'gray'

# iterate launches, create marker + popup
for _, r in spacex_df.iterrows():
    # try common column names
    lat = r.get('Lat') or r.get('Latitude') or r.get('lat')
    lon = r.get('Long') or r.get('Longitude') or r.get('lon')
    if pd.isna(lat) or pd.isna(lon):
        continue
    coord = [float(lat), float(lon)]

    # popup content (short)
    popup = folium.Popup(
        f"Site: {r.get('LaunchSite', r.get('Launch Site', 'N/A'))}<br>"
        f"Date: {r.get('Date','N/A')}<br>"
        f"Orbit: {r.get('Orbit','N/A')}<br>"
        f"Payload(kg): {r.get('PayloadMass', r.get('PAYLOAD_MASS__KG_', 'N/A'))}<br>"
        f"Outcome: {'Success' if str(r.get('class','0'))=='1' else 'Failure'}",
        max_width=300
    )

    folium.CircleMarker(
        location=coord,
        radius=5,
        color=color_by_class(r.get('class')),
        fill=True,
        fill_opacity=0.8,
        popup=popup
    ).add_to(launch_cluster)

# display and save
site_map.save("spacex_launch_all_markers.html")
display(HTML(site_map._repr_html_()))
print("Saved: spacex_launch_all_markers.html")
```

```python
from IPython.core.display import HTML, display
from folium.features import DivIcon
try:
    marker_cluster
except NameError:
    from folium.plugins import MarkerCluster
    marker_cluster = MarkerCluster().add_to(site_map)
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# For each row in spacex_df, create a Marker and add it to the cluster
for index, record in spacex_df.iterrows():
    # attempt common latitude/longitude column names
    lat = record.get('Lat') or record.get('Latitude') or record.get('lat')
    lon = record.get('Long') or record.get('Longitude') or record.get('lon')
    if pd.isna(lat) or pd.isna(lon):
        continue

    coord = [float(lat), float(lon)]

    # popup (short)
    site_name = record.get('LaunchSite') or record.get('Launch Site') or record.get('Site') or 'Launch Site'
    popup_html = (f"Site: {site_name}<br>"
                  f"Date: {record.get('Date','N/A')}<br>"
                  f"Orbit: {record.get('Orbit','N/A')}<br>"
                  f"Payload(kg): {record.get('PayloadMass', record.get('PAYLOAD_MASS__KG_', 'N/A'))}<br>"
                  f"Outcome: {'Success' if str(record.get('class','0'))=='1' else 'Failure'}")

    # create marker with custom icon color indicator (uses marker_color column)
    marker = folium.Marker(
        location=coord,
        popup=folium.Popup(popup_html, max_width=300),
        icon=folium.Icon(color='white', icon_color=record.get('marker_color', 'gray'))
    )

    # add to cluster
    marker_cluster.add_child(marker)

# Show the map (use the method that worked for you)
display(HTML(site_map._repr_html_()))
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
from folium.plugins import MousePosition
from IPython.core.display import HTML, display

# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
display(HTML(site_map._repr_html_()))
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from folium.features import DivIcon

# --- Crear un mapa nuevo centrado en el launch site ---
launch_site_lat = 28.561857
launch_site_lon = -80.577366

site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# (Opcional) marcar el launch site
launch_marker = folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color="green")
)
site_map.add_child(launch_marker)

# --- Coordenadas reales ---
titan_lat = 28.56213     # Titan III Road
titan_lon = -80.56777

coastline_lat = 28.56229 # Coastline ajustado
coastline_lon = -80.58733

# --------- Launch Site  Coastline ---------
distance_coast = calculate_distance(
    launch_site_lat, launch_site_lon,
    coastline_lat, coastline_lon
)

line_coast = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    color="orange",
    weight=2
)
site_map.add_child(line_coast)

marker_coast = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(150, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>Titan III road: {:.2f} KM</b></div>'.format(distance_coast)
    )
)
site_map.add_child(marker_coast)

# --------- Launch Site  Titan III Road ---------
distance_titan = calculate_distance(
    launch_site_lat, launch_site_lon,
    titan_lat, titan_lon
)

line_titan = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [titan_lat, titan_lon]],
    color="blue",
    weight=2
)
site_map.add_child(line_titan)

marker_titan = folium.Marker(
    [titan_lat, titan_lon],
    icon=DivIcon(
        icon_size=(150, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#1f4ba8;"><b>Coast line: {:.2f} KM</b></div>'.format(distance_titan)
    )
)
site_map.add_child(marker_titan)

# Mostrar mapa
site_map
```
